```python
fig, ax = plt.subplots(figsize=(20,10))

font = FontProperties()
font.set_name('Times New Roman')
plt.rcParams['xtick.labelsize'] = 70
plt.rcParams['ytick.labelsize'] = 30
plt.rcParams['figure.dpi'] = 300

Amp_1MHz = [0.7708333333333334,0.7916666666666666, 0.8125,0.875,0.9166666666666666,0.9166666666666666]

x_position = np.arange(6)
cell_types = ['KNN','LR', 'DT', 'RF', 'SVM', 'ANN']

ax.bar(x_position, Amp_1MHz,  align='center', alpha=0.5, width = 0.7, ecolor='black', capsize=10, color = ['r','g','b','c','m','y'])
ax.set_ylabel('Accuracy', fontsize=30,  fontweight='bold')
ax.set_xticks(x_position)
ax.set_xticklabels(cell_types)
ax.set_title('Models', y=-0.22, fontsize=40, fontweight='bold')
ax.yaxis.grid(False)


bottom = 0.7
top = 1

ax.set_ylim(bottom, top)
plt.show()
```
